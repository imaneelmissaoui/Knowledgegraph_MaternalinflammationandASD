```python
import pandas as pd
```

```python
from metapub import PubMedFetcher
fetch = PubMedFetcher()

# get the  PMID for first 3 articles with keyword sepsis
pmids = ["35131181","28455196","36300375"]

# get  articles
articles = {}
for pmid in pmids:
    articles[pmid] = fetch.article_by_pmid(pmid)
```

```python
# get title for each article:
titles = {}
for pmid in pmids:
    titles[pmid] = fetch.article_by_pmid(pmid).title
Title = pd.DataFrame(list(titles.items()),columns = ['pmid','Title'])
Title
```

```python
# get abstract for each article:
abstracts = {}
for pmid in pmids:
    abstracts[pmid] = fetch.article_by_pmid(pmid).abstract
Abstract = pd.DataFrame(list(abstracts.items()),columns = ['pmid','Abstract'])
Abstract
```

```python
# get author for each article:
authors = {}
for pmid in pmids:
    authors[pmid] = fetch.article_by_pmid(pmid).authors
Author = pd.DataFrame(list(authors.items()),columns = ['pmid','Author'])
Author

```

```python
# get year for each article:
years = {}
for pmid in pmids:
    years[pmid] = fetch.article_by_pmid(pmid).year
Year = pd.DataFrame(list(years.items()),columns = ['pmid','Year'])
Year

```

```python
# get volume for each article:
volumes = {}
for pmid in pmids:
    volumes[pmid] = fetch.article_by_pmid(pmid).volume
Volume = pd.DataFrame(list(volumes.items()),columns = ['pmid','Volume'])
Volume
```

```python
# get issue for each article:
issues = {}
for pmid in pmids:
    issues[pmid] = fetch.article_by_pmid(pmid).issue
Issue = pd.DataFrame(list(issues.items()),columns = ['pmid','Issue'])
Issue
```

```python
# get journal for each article:
journals = {}
for pmid in pmids:
    journals[pmid] = fetch.article_by_pmid(pmid).journal
Journal = pd.DataFrame(list(journals.items()),columns = ['pmid','Journal'])
Journal
```

```python
# get citation for each article:
citations = {}
for pmid in pmids:
    citations[pmid] = fetch.article_by_pmid(pmid).citation
Citation = pd.DataFrame(list(citations.items()),columns = ['pmid','Citation'])
Citation
```

```python
links={}
for pmid in pmids:
    links[pmid] = "https://pubmed.ncbi.nlm.nih.gov/"+pmid+"/"
Link = pd.DataFrame(list(links.items()),columns = ['pmid','Link'])
Link
```

```python
#get full text of each article
import metapub
from urllib.request import urlretrieve
import textract

for pmid in pmids:
    try:
        # Find URL of full-text article
        url = metapub.FindIt(pmid).url

        # Download PDF of full-text article
        pdf_path = f'{pmid}.pdf'  # Naming convention for saving PDFs
        urlretrieve(url, pdf_path)

        # Convert PDF to text
        text = textract.process(pdf_path, extension='pdf', method='pdftotext', encoding='utf_8')

        # Save text to file
        txt_path = f'{pmid}.txt'  # Naming convention for saving text files
        with open(txt_path, 'w', encoding='utf-8') as textfile:
            textfile.write(text.decode('utf-8'))

        print(f"Full text for PMID {pmid} retrieved and saved successfully.")

    except Exception as e:
        print(f"Error processing PMID {pmid}: {str(e)}")
```

```python
import os
from tqdm import tqdm
from preprocess import text_to_batches

def process_articles(base_path):
    # Create the 'data' folder if it doesn't exist
    data_folder = os.path.join(base_path, 'data')
    if not os.path.exists(data_folder):
        os.makedirs(data_folder)

    # Create the 'cleaned' folder if it doesn't exist
    cleaned_folder = os.path.join(base_path, 'cleaned')
    if not os.path.exists(cleaned_folder):
        os.makedirs(cleaned_folder)
    # Process articles
    for file_name in tqdm(os.listdir(base_path)):
        if not file_name.endswith('.txt'):
            continue  # Skip non-text files

        # Check if results already processed
        if os.path.exists(os.path.join(data_folder, file_name)):
            continue
        results = ''
        with open(os.path.join(base_path, file_name), 'r', encoding='utf-8') as f:
            content = f.read()
            batches = text_to_batches(content)  # Split text into batches
            # Process each batch through ChatGPT 3.5
            for batch in batches:
                gpt_results = process_gpt3_5(batch)  # Process text using ChatGPT 3.5
                results += gpt_results

        # Save outputs from model in 'data' folder
        with open(os.path.join(data_folder, file_name), 'w', encoding='utf-8') as results_file:
            results_file.write(results)

        # Save actual texts in 'cleaned' folder
        with open(os.path.join(cleaned_folder, file_name), 'w', encoding='utf-8') as cleaned_file:
            cleaned_file.write(content)

# Mount Google Drive
drive.mount('/content/drive')
# Call the function with the base path to the 'raw' folder
base_path = "/content/drive/MyDrive/raw"
process_articles(base_path)
```
